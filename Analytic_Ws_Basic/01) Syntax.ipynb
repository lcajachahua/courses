```
SELECT customer_id, page, datestamp FROM TRNG_TDU_TD01.bank_web;
```

```
-- Advanced SQL Engine Query (without alias) 
SELECT * FROM Sessionize 
(ON TRNG_TDU_TD01.bank_web 
 PARTITION BY customer_id 
 ORDER BY datestamp 
 USING
 TimeColumn ('datestamp')
 TimeOut (600)
 );
```

```
-- Advanced SQL Engine Query (with alias) 
SELECT * FROM Sessionize 
(ON TRNG_TDU_TD01.bank_web 
 PARTITION BY customer_id 
 ORDER BY datestamp 
 USING
 TimeColumn ('datestamp')
 TimeOut (600)
) AS dt;
```

```
--------------------------------------------------
-- Lab 02: Function Argument Order --
-- Change order of arguments after USING
--------------------------------------------------
SELECT * FROM TRNG_TDU_TD01.borre_z;

SELECT * FROM nPath 
(ON TRNG_TDU_TD01.borre_z
 PARTITION BY user_id 
 ORDER BY ts
 USING
 Symbols (event = 'a' AS X)
 Pattern ('X.X')
 Mode (NONOVERLAPPING)
  Result (Accumulate (event OF X) AS x_pattern)
) AS dt;
```

This code emulates the partition by and order by

```
SELECT * FROM TRNG_TDU_TD01.table1 ORDER BY 1,3;
```

```
--------------------------------------------------
-- Lab 03: Partition By and Order By --
--------------------------------------------------
-- ANSI SQL
SELECT * FROM TRNG_TDU_TD01.table1 ORDER BY 3;

-- PARTITION BY and ORDER BY
SELECT * FROM Sessionize 
(ON TRNG_TDU_TD01.table1
 PARTITION BY id
 ORDER BY ts 
 USING
 TimeColumn ('ts')
 TimeOut (5)
) ORDER BY id, ts;
```

```
----------------------------------------
-- Lab 04: Dimension Tables --
----------------------------------------
-- View 3 tables
SELECT * FROM TRNG_TDU_TD01.borre_y;
SELECT * FROM TRNG_TDU_TD01.conv_table;
SELECT * FROM TRNG_TDU_TD01.model1_table ORDER BY 1;
```

```
EXPLAIN
SELECT * FROM Attribution
(ON TRNG_TDU_TD01.borre_y PARTITION BY user_id ORDER BY ts
 ON TRNG_TDU_TD01.conv_table AS conversion DIMENSION
 ON TRNG_TDU_TD01.model1_table AS model1 DIMENSION 
 USING
 EventColumn ('event')
 TimestampColumn ('ts')
 WindowSize ('rows:8')
) AS dt ORDER BY user_id, ts;
```
